```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path_csv = 'BCI Aug 2024.csv'
csv_data = pd.read_csv(file_path_csv)

# Clean the data by setting the correct header and parsing dates
csv_data_cleaned = pd.read_csv(file_path_csv, skiprows=1, names=['Date', 'Leading Indicator', 'Coincident Indicator', 'Lagging Indicator'])

# Convert 'Date' column to datetime
csv_data_cleaned['Date'] = pd.to_datetime(csv_data_cleaned['Date'], errors='coerce')

# Drop rows with NaT in 'Date' column
csv_data_cleaned = csv_data_cleaned.dropna(subset=['Date'])

# Convert the remaining columns to numeric, coercing errors and dropping non-numeric rows
for col in ['Leading Indicator', 'Coincident Indicator', 'Lagging Indicator']:
    csv_data_cleaned[col] = pd.to_numeric(csv_data_cleaned[col], errors='coerce')

# Drop any rows where all indicator columns are NaN
csv_data_cleaned = csv_data_cleaned.dropna(how='all', subset=['Leading Indicator', 'Coincident Indicator', 'Lagging Indicator'])

# Plot the data with gridlines
plt.figure(figsize=(12, 6))
plt.plot(csv_data_cleaned['Date'], csv_data_cleaned['Leading Indicator'], label='Leading Indicator')
plt.plot(csv_data_cleaned['Date'], csv_data_cleaned['Coincident Indicator'], label='Coincident Indicator')
plt.plot(csv_data_cleaned['Date'], csv_data_cleaned['Lagging Indicator'], label='Lagging Indicator')

# Set plot title and labels
plt.title('Economic Indicators Over Time (Indexed to 2019=100)')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.legend()

# Add gridlines
plt.grid(True)

# Display the plot
plt.show()

# Generate summary statistics for the indicators
summary_statistics = csv_data_cleaned[['Leading Indicator', 'Coincident Indicator', 'Lagging Indicator']].describe()

# Print the summary statistics
print("Summary Statistics for Economic Indicators:")
print(summary_statistics)
```

Interpreting economic indicators like the Leading, Coincident, and Lagging Indicators is crucial for understanding the state of the economy and predicting future economic activity. Here's how to interpret each:

1. Leading Indicator:
Purpose: The Leading Indicator is designed to predict future economic activity. It typically changes before the economy starts to follow a particular trend, making it useful for forecasting.
Examples: Stock market performance, new business startups, consumer confidence, and new orders for durable goods.
Interpretation:
Increase: An upward trend in the Leading Indicator suggests that the economy might expand in the near future. It indicates growing confidence, higher spending, and investment.
Decrease: A downward trend might signal an impending slowdown or recession. It suggests that businesses and consumers might be holding back on spending and investments.
2. Coincident Indicator:
Purpose: The Coincident Indicator reflects the current state of the economy. It moves in line with the overall economic cycle, providing a real-time snapshot of economic conditions.
Examples: Employment levels, personal income, industrial production, and retail sales.
Interpretation:
Increase: A rising Coincident Indicator suggests that the economy is currently expanding. It indicates strong job growth, higher production, and increased consumer spending.
Decrease: A decline signals that the economy might be contracting. It could be a sign of increasing unemployment, lower production, and reduced spending.
3. Lagging Indicator:
Purpose: The Lagging Indicator typically changes after the economy has already begun to follow a particular trend. It confirms patterns and trends but does not predict them.
Examples: Unemployment rates, labor costs, consumer debt, and interest rates.
Interpretation:
Increase: An increase in the Lagging Indicator might indicate that the economy has been performing well for a while, but it may also suggest that inflationary pressures or rising costs are becoming a concern.
Decrease: A decrease could mean that the economy has been slowing down, but it could also indicate that inflation and costs are stabilizing after a period of growth.
How They Work Together:
Forecasting and Decision Making: Analysts and policymakers use these indicators together to get a comprehensive view of the economy. The Leading Indicator helps forecast future economic conditions, the Coincident Indicator provides insight into the current state of the economy, and the Lagging Indicator confirms long-term trends and potential shifts.
Business Cycle Analysis: These indicators are critical for identifying different phases of the business cycleexpansion, peak, contraction, and trough. By observing the trends in these indicators, economists can make informed predictions about where the economy is headed and help businesses and governments prepare accordingly.
Example Interpretation:
If the Leading Indicator is rising, the Coincident Indicator is steady, and the Lagging Indicator is just beginning to rise, it might suggest that the economy is in the early stages of expansion. Conversely, if the Leading Indicator is falling, the Coincident Indicator is starting to decline, and the Lagging Indicator is still rising, it could indicate that the economy is entering a recessionary phase.
Understanding and analyzing these indicators helps businesses, investors, and policymakers make informed decisions about spending, investment, and economic policy.
