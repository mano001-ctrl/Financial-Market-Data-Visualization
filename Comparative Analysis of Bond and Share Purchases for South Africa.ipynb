```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'MonthlyRelease_Capital_Markets.csv'
data = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%b, %Y')

# Convert relevant columns to numeric, handling errors and removing commas if present
numeric_columns = [
    'Value of shares traded (Rm)',
    'Number of shares traded (millions)',
    'Net purchases of shares',
    'Net purchases of bonds',
    'Total net purchases of shares and bonds',
    'Net issues of marketable debt securities (Rm)',
    'Nominal value of bonds traded (Rm)',
    'Consideration of bonds traded (Rm)',
    'Number of transactions (actual number)',
    'Yield on long-term government bonds (%)',
    'Number of deals (actual number)',
    'Number of contracts (actual number)',
    'Underlying value (Rm)',
    'Open interest (actual number)'
]

columns_to_convert = ['Number of deals (actual number)', 
                      'Number of contracts (actual number)', 
                      'Underlying value (Rm)', 
                      'Open interest (actual number)']

# Convert columns to numeric where needed
for column in numeric_columns + columns_to_convert:
    if data[column].dtype == 'object':  # Check if the column is of type object
        data[column] = pd.to_numeric(data[column].str.replace(',', ''), errors='coerce')
    else:
        data[column] = pd.to_numeric(data[column], errors='coerce')

# 1. Plot Net Purchases of Bonds Over Time
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Net purchases of bonds'], marker='o')
plt.title('Net Purchases of Bonds Over Time')
plt.xlabel('Date')
plt.ylabel('Net Purchases of Bonds (Rm)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Plot Net Purchases of Shares Over Time
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Net purchases of shares'], marker='o', color='orange')
plt.title('Net Purchases of Shares Over Time')
plt.xlabel('Date')
plt.ylabel('Net Purchases of Shares (Rm)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Comparison of Net Purchases: Bonds vs. Shares Over Time
plt.figure(figsize=(12, 7))
plt.plot(data['Date'], data['Net purchases of bonds'], marker='o', label='Net Purchases of Bonds', color='blue')
plt.plot(data['Date'], data['Net purchases of shares'], marker='o', label='Net Purchases of Shares', color='orange')
plt.title('Comparison of Net Purchases: Bonds vs. Shares Over Time')
plt.xlabel('Date')
plt.ylabel('Net Purchases (Rm)')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# 6. Cumulative Net Purchases of Bonds
data['Cumulative Net Purchases of Bonds'] = data['Net purchases of bonds'].cumsum()
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Cumulative Net Purchases of Bonds'], marker='o', color='purple')
plt.title('Cumulative Net Purchases of Bonds Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Net Purchases (Rm)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 7. Monthly Net Purchases of Bonds
plt.figure(figsize=(10, 6))
plt.bar(data['Date'], data['Net purchases of bonds'], color='skyblue')
plt.title('Monthly Net Purchases of Bonds')
plt.xlabel('Date')
plt.ylabel('Net Purchases of Bonds (Rm)')
plt.grid(True, axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

1. Net Purchases of Bonds Over Time
Description: This line chart illustrates the monthly net purchases of bonds over time.
Commentary: The net purchases of bonds show significant fluctuations, with periods of both buying (positive values) and selling (negative values). These movements can be attributed to various factors such as changes in interest rates, economic conditions, or shifts in investor sentiment. For instance, spikes in bond purchases might coincide with market volatility or economic downturns, where investors seek safer assets. Conversely, periods of selling may indicate a move towards riskier investments or expectations of rising interest rates.
2. Net Purchases of Shares Over Time
Description: This line chart tracks the monthly net purchases of shares over time.
Commentary: The net purchases of shares also exhibit variability, reflecting market dynamics and investor behavior. Positive values represent net buying activity, often occurring during bullish market conditions when confidence in equities is high. On the other hand, negative values indicate net selling, which could be driven by market corrections, profit-taking, or shifts towards safer investments. The chart helps to identify periods of heightened activity in the stock market and can be used to analyze how external events influence investor decisions.
3. Comparison of Net Purchases: Bonds vs. Shares Over Time
Description: This comparative line chart shows the net purchases of bonds and shares over time on the same axis.
Commentary: The chart allows for a direct comparison between investor activity in bonds and shares. In some periods, you may observe an inverse relationship, where net purchases of bonds increase as net purchases of shares decrease, indicating a shift from equities to bonds, possibly due to risk aversion or market uncertainty. Conversely, simultaneous increases or decreases in both asset classes could suggest broad market trends, such as overall optimism or pessimism affecting all types of investments. This chart provides valuable insights into how investors allocate their portfolios between these two major asset classes.
6. Cumulative Net Purchases of Bonds Over Time
Description: This cumulative line chart shows the total accumulation of net bond purchases over time.
Commentary: The cumulative chart offers a long-term perspective on bond investments, smoothing out the month-to-month volatility seen in the previous charts. A consistently rising line indicates sustained buying over time, reflecting overall confidence in bonds as an investment. Plateaus or declines in the cumulative sum might correspond to periods of market uncertainty, shifts in monetary policy, or changing economic conditions that prompt investors to reduce their bond holdings. This chart is particularly useful for understanding the overall trend in bond market investments over an extended period.
7. Monthly Net Purchases of Bonds
Description: This bar chart provides a monthly breakdown of net bond purchases.
Commentary: The bar chart highlights the month-to-month variation in bond purchases, making it easier to spot specific months with unusually high or low activity. Tall positive bars indicate strong net buying activity, possibly due to factors like declining interest rates, which make bonds more attractive. Conversely, tall negative bars signal net selling, which could occur in response to rising interest rates or a preference for riskier assets. The clear visual distinctions provided by this chart allow for a detailed examination of short-term trends in bond purchases.
