```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (update the path if needed)
file_path = '90-day Rates and Yields Treasury Securities for the United Kingdom.csv'
data = pd.read_csv(file_path)

# Convert DATE column to datetime format
data['DATE'] = pd.to_datetime(data['DATE'])

# Calculate volatility as the rolling standard deviation (12-month window)
data['Volatility'] = data['IR3TTS01GBM156N'].rolling(window=12).std()

# Calculate the moving average with a 12-month window
data['Moving_Average'] = data['IR3TTS01GBM156N'].rolling(window=12).mean()

# Plot the original data
plt.figure(figsize=(10, 6))
plt.plot(data['DATE'], data['IR3TTS01GBM156N'], label='90-day Rate & Yield')
plt.title('90-day Rates and Yields for UK Treasury Securities')
plt.xlabel('Date')
plt.ylabel('Rate/Yield (%)')
plt.grid(True)  # Add gridlines
plt.legend()
plt.show()

# Plot the volatility
plt.figure(figsize=(10, 6))
plt.plot(data['DATE'], data['Volatility'], label='Volatility (Rolling Std Dev)', color='orange')
plt.title('Volatility of 90-day Rates and Yields for UK Treasury Securities')
plt.xlabel('Date')
plt.ylabel('Volatility (%)')
plt.grid(True)  # Add gridlines
plt.legend()
plt.show()

# Plot the moving average
plt.figure(figsize=(10, 6))
plt.plot(data['DATE'], data['Moving_Average'], label='Moving Average (12-month)', color='green')
plt.title('Moving Average of 90-day Rates and Yields for UK Treasury Securities')
plt.xlabel('Date')
plt.ylabel('Rate/Yield (%)')
plt.grid(True)  # Add gridlines
plt.legend()
plt.show()

# Additional plot: Difference between the current and previous rate (rate of change)
data['Rate_of_Change'] = data['IR3TTS01GBM156N'].diff()

plt.figure(figsize=(10, 6))
plt.plot(data['DATE'], data['Rate_of_Change'], label='Rate of Change', color='red')
plt.title('Rate of Change in 90-day Rates and Yields for UK Treasury Securities')
plt.xlabel('Date')
plt.ylabel('Rate of Change (%)')
plt.grid(True)  # Add gridlines
plt.legend()
plt.show()
```

Here’s a detailed commentary for each plot generated from the data of the 90-day UK Treasury securities:

1. Original Data: 90-day Rates and Yields for UK Treasury Securities
Plot Overview: This plot shows the time series of 90-day Treasury bill rates and yields from 1960 onward.
Key Observations:
The rates exhibit clear cyclical behavior with noticeable peaks and troughs.
During periods of economic uncertainty or high inflation, such as the late 1970s and early 1980s, the rates sharply increase, reflecting higher demand for returns to offset inflation.
The late 1990s and post-2008 financial crisis periods show significantly lower rates, which is typical of a low-interest-rate environment driven by central banks attempting to stimulate economic growth.
Recent years (from the 2000s onward) show persistently lower yields compared to earlier periods, consistent with global trends of lower interest rates and yields.
2. Volatility of 90-day Rates and Yields
Plot Overview: Volatility is computed as the rolling standard deviation over a 12-month window, representing the rate’s fluctuation intensity.
Key Observations:
The most volatile period occurs during the 1970s and early 1980s, which coincides with high inflation and economic instability (stagflation) in the UK and globally. This is a period of sharp increases and decreases in interest rates as central banks intervened to curb inflation.
Volatility is notably low after 1990, reflecting more stable economic conditions and central bank interventions to smooth out economic cycles.
Post-2008, the volatility stays subdued, highlighting the "lower for longer" interest rate regime introduced by central banks to support economies after the financial crisis and during the COVID-19 pandemic.
A recent uptick in volatility might suggest rising uncertainties in financial markets, potentially linked to global economic events.
3. Moving Average of 90-day Rates and Yields (12-month)
Plot Overview: The moving average smooths out short-term fluctuations to reveal longer-term trends.
Key Observations:
The moving average highlights long-term trends in interest rates and yields, filtering out month-to-month noise.
Peaks are prominent during the late 1970s and early 1980s, confirming that this was a high-interest-rate period driven by inflationary pressure.
The 1990s through the 2000s are characterized by a prolonged decline in yields, reflecting global trends of falling interest rates as central banks shifted towards more accommodative monetary policies.
The post-2008 period, particularly after the global financial crisis, shows a steady decline in yields, with the moving average remaining flat at historically low levels. This is indicative of a prolonged low-rate environment, with central banks like the Bank of England cutting rates to near zero.
The moving average also smooths out any temporary rate spikes, making it useful for identifying periods of secular decline or rise.
4. Rate of Change in 90-day Rates and Yields
Plot Overview: This plot shows the difference in consecutive 90-day yields, which helps identify how quickly the rates are increasing or decreasing over time.
Key Observations:
The rate of change plot is highly volatile, with spikes both positive and negative. This reflects sudden and sometimes extreme adjustments in monetary policy, particularly in response to inflationary shocks or economic crises.
Large positive changes during the 1970s and early 1980s indicate aggressive hikes in interest rates in response to runaway inflation.
There are also sharp negative spikes during times of economic downturn, such as the early 1980s recession and the 2008 financial crisis, when rates were cut dramatically to stimulate growth.
The more recent period, especially post-2008, shows less pronounced fluctuations in the rate of change, reflecting a more stable, lower-rate environment where central banks have moved cautiously with incremental adjustments.
Comprehensive Analysis:
Economic Context: The data and all the plots together provide a clear picture of the UK’s monetary policy trends and the broader economic environment over the past several decades. The high volatility and large rate changes during the 1970s-1980s reflect the inflationary pressures and economic volatility, while the post-2000 period is marked by subdued rates and low volatility, characteristic of a low-interest-rate world post-financial crisis.

Key Time Periods:

Late 1960s to 1980s: Marked by high interest rates and high volatility, driven by inflation and central bank actions to curb it.
1990s: Characterized by a gradual decline in rates and volatility as inflation was brought under control.
Post-2008: A period of historically low rates and yields, with very little volatility, driven by central banks' actions to support weak economies after the financial crisis.
Investment Insight: For fixed-income investors, this analysis reveals that long periods of low yields (like post-2008) may pose challenges in finding returns. High volatility periods, though risky, also offer higher return potential, but they come with the need for more active risk management.

These insights offer a comprehensive look at the interest rate environment in the UK over time and provide useful context for future rate expectations and investment strategy planning.
